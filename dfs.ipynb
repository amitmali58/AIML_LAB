```python
def dfs(graph, start, goal):
    visited, path = set(), []

    def explore(node):
        if node in visited:
            return False
        visited.add(node)
        path.append(node)
        if node == goal:
            return True
        for nxt in graph[node]:
            if explore(nxt):
                return True
        path.pop()
        return False

    explore(start)
    return list(visited), path


# Simple graph
graph = {
    "Home": ["Park", "Mall"],
    "Park": ["Lake"],
    "Mall": ["Cinema"],
    "Lake": [],
    "Cinema": []
}

start, goal = "Home", "Cinema"
order, path = dfs(graph, start, goal)

for i, n in enumerate(order, 1):
    print(f"{i}. {n}")
if path and path[-1] == goal:
    print("\nPath to goal:", " -> ".join(path))
else:
    print("\nGoal not found!")
```
